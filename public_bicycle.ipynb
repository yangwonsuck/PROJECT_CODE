```python
df['datetime']=df['datetime'].str.replace(pat=' ',repl='-') # hourly date 와 timestamp 사이의 빈칸을 -로 대체
df['datetime']=df['datetime'].str.replace(pat=':',repl='-') # timestamp의 :구분자를 -로 대체
df['datetime'].str.split('-') #yyyy-MM-dd-hh-mm -ss를 나누는 구분자인 -를 기준으로 분리
```

```python
df.head()
```

```python
import datetime # 날짜를 추출하기 위한 라이브러리 선언
df['year']=pd.DatetimeIndex(df['datetime']).year  #연도를 따로 뽑아냄 
df['month']=pd.DatetimeIndex(df['datetime']).month #월을 따로 뽑아냄
df['day']=pd.DatetimeIndex(df['datetime']).day #날짜를 따로 뽑아냄
df['hour']=pd.DatetimeIndex(df['datetime']).hour #시간을 따로 뽑아냄
df['minute']=pd.DatetimeIndex(df['datetime']).minute #분을 따로 뽑아냄
df['second']=pd.DatetimeIndex(df['datetime']).second #초를 따로 뽑아냄
df['dayofweek']=pd.DatetimeIndex(df['datetime']).dayofweek #요일을 따로 뽑아냄
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(20,7))
plt.subplot(2,3,1)
plt.title('year')
plt.plot(df['year'],df['count']) #연도와 대여수를 시각화 -> 2011초반과 2012년 초반이 대여수가 급등함
plt.subplot(2,3,2)
plt.title('month')
plt.plot(df['month'],df['count'],'r') #달과 대여수를 시각화 ->여름에서 가을 사이가 대여수가 많음
plt.subplot(2,3,3)
plt.title('day')
plt.plot(df['day'],df['count'],'b') #일자와 대여수를 시각화 
plt.subplot(2,3,4)
plt.title('hour')
plt.plot(df['hour'],df['count'],'g') #시간과 대여수를 시각화 ->오전 7~8시와 12~13시 17~18 시 사이가 대여수가 많음
plt.subplot(2,3,5)
plt.title('minute')
plt.plot(df['minute'],df['count']) #분과 대여수를 시각화
plt.subplot(2,3,6)
plt.title('second') 
plt.plot(df['second'],df['count']) #초와 대여수를 시각화
plt.show()
```

```python
#전체 대여량, 계절별 대여량, 시간별 대여량, 근무일에 따른 대여량

plt.figure(figsize=(20,7))
plt.subplot(2,3,1)
plt.title('total_count')
plt.plot(df['count']) #전체대여수를 시각화 -> 점점 늘어나게 됨

plt.subplot(2,3,2)
plt.title('season')
plt.plot(df['season'],df['count'],'r') #계절과 대여수를 시각화 -> 여름과 가을에 대여수가 급등함

plt.subplot(2,3,3)
plt.title('hour')
plt.plot(df['hour'],df['count'],'g') #시간과 대여수를 시각화 ->오전 7~8시와 12~13시 17~18 시 사이가 대여수가 많음

plt.subplot(2,3,4)
plt.title('workingday')
plt.plot(df['workingday'],df['count'],'b') #일하는 날과대여수를 시각화 -> 일을 하는 날이 대여수가 더 많음
```
